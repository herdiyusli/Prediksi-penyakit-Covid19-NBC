```python
import pandas as pd
import numpy as np
```

```python
dataset =  pd.read_csv("covid19.csv")
dataset.head()
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isna().sum()
```

```python
dataset.replace('YES',1,inplace = True)
dataset.replace('NO',0,inplace = True)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# Variabel independen
x = dataset.drop(["Terinfeksi-COVID"], axis = 1)
# Variabel dependen
y = dataset["Terinfeksi-COVID"]
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_test
```

```python
from sklearn.naive_bayes import CategoricalNB
classifier = CategoricalNB()
classifier.fit(x_train, y_train)
classifier.class_count_
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
classifier.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
import seaborn as sns
group_names = ['True Neg','False Neg','False Pos','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
```

```python
print("Akurasi dari model Naive Bayes Classifier = ",classifier.score(x_test,y_test)*100,"%")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
